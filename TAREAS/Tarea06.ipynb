1. Determine el orden de la mejor aproximación para las siguientes funciones, usando la Serie de Taylor y el Polinomio de Lagrange:

$$ \frac{1}{25*x^2+1} , x0 = 0$$

Derivadas calculadas en x₀ = 0:

$$ f(0) = 1

f'(0) = 0

f''(0) = -50

f'''(0) = 0

f⁴(0) = 7500

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 500)

# Función real
f1 = 1 / (25*x**2 + 1)

# orden 2
P2 = 1 - 25*x**2
print("Orden 2: P₂(x) = 1 - 25x²")

# orden 4
P4 = 1 - 25*x**2 + 625*x**4
print("Orden 4: P₄(x) = 1 - 25x² + 625x⁴")

plt.figure(figsize=(10, 6))
plt.plot(x, f1, 'k-', linewidth=2.5, label='f(x) = 1/(25x²+1)')
plt.plot(x, P2, 'b--', linewidth=2, label='Orden 2')
plt.plot(x, P4, 'r--', linewidth=2, label='Orden 4')
plt.plot(0, 1, 'go', markersize=10, label='x₀=0')
plt.grid(True, alpha=0.3)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ejercicio 1: Serie de Taylor')
plt.legend()
plt.ylim(-0.5, 1.5)
plt.show()
```

```python
print("  Orden 2: P₂(x) = 1 - 25x²")
print("  Orden 4: P₄(x) = 1 - 25x² + 625x⁴")
```

---

$$ arctangx, x0 = 1 $$

Derivadas en x₀ = 1:

$$ f(1) = π/4 ≈ 0.7854

f'(1) = 1/2 = 0.5

f''(1) = -1/2 = -0.5

f'''(1) = 1/2 = 0.5

```python
x = np.linspace(-1, 3, 500)

# Función real
f2 = np.arctan(x)

# orden 1
P1 = np.pi/4 + 0.5*(x - 1)
print(f"Orden 1: P₁(x) = π/4 + 0.5(x-1)")

# orden 2
P2 = np.pi/4 + 0.5*(x - 1) - 0.25*(x - 1)**2
print(f"Orden 2: P₂(x) = π/4 + 0.5(x-1) - 0.25(x-1)²")

# orden 3
P3 = np.pi/4 + 0.5*(x - 1) - 0.25*(x - 1)**2 + (1/12)*(x - 1)**3
print(f"Orden 3: P₃(x) = π/4 + 0.5(x-1) - 0.25(x-1)² + (1/12)(x-1)³")

plt.figure(figsize=(10, 6))
plt.plot(x, f2, 'k-', linewidth=2.5, label='f(x) = arctan(x)')
plt.plot(x, P1, 'b--', linewidth=2, label='Orden 1')
plt.plot(x, P2, 'g--', linewidth=2, label='Orden 2')
plt.plot(x, P3, 'r--', linewidth=2, label='Orden 3')
plt.plot(1, np.pi/4, 'mo', markersize=10, label='x₀=1')
plt.grid(True, alpha=0.3)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ejercicio 2: Serie de Taylor')
plt.legend()
plt.show()
```

```python
print("  Orden 1: P₁(x) = π/4 + 0.5(x-1)")
print("  Orden 2: P₂(x) = π/4 + 0.5(x-1) - 0.25(x-1)²")
print("  Orden 3: P₃(x) = π/4 + 0.5(x-1) - 0.25(x-1)² + (1/12)(x-1)³")
```
