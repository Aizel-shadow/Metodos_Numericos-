#### Ejercicio 4: Aritmética de redondeo de 3 dígitos

Usar aritmética de redondeo a 3 cifras significativas y calcular errores.

1. Redondear en cada operación a 3 cifras significativas
2. Operar de izquierda a derecha
3. Usar notación de punto flotante $fl(x)$

```python
def fl(x, digits=3):
    """
    Función para redondeo a n cifras significativas
    Simula la función fl(x) del material de clase
    """
    if x == 0:
        return 0
    from math import log10, floor
    return round(x, -int(floor(log10(abs(x)))) + (digits - 1))

# Verificar la función
print(f"fl(0.928571, 3) = {fl(0.928571, 3)}")
print(f"fl(3.14159, 3) = {fl(3.14159, 3)}")
print(f"fl(0.0492, 3) = {fl(0.0492, 3)}")
```

a) $\frac{13}{14} - \frac{5}{7} + 2e - 5.4$

```python
print("Expresión: 13/14 - 5/7 + 2e - 5.4\n")

# Paso 1: Convertir a punto flotante
val1 = fl(13/14, 3)
val2 = fl(5/7, 3)
val3 = fl(2*np.e, 3)
val4 = fl(5.4, 3)

print(f"fl(13/14) = {val1}")
print(f"fl(5/7) = {val2}")
print(f"fl(2e) = {val3}")
print(f"fl(5.4) = {val4}\n")

# Paso 2: Operaciones de izquierda a derecha
result1 = fl(val1 - val2, 3)
print(f"Paso 1: fl({val1} - {val2}) = {result1}")

result2 = fl(result1 + val3, 3)
print(f"Paso 2: fl({result1} + {val3}) = {result2}")

result3 = fl(result2 - val4, 3)
print(f"Paso 3: fl({result2} - {val4}) = {result3}")

print(f"\nResultado con aritmética de 3 dígitos: {result3}")

# Valor exacto
exact_value = 13/14 - 5/7 + 2*np.e - 5.4
print(f"Valor exacto: {exact_value:.10f}")

# Errores
error_abs_4a = abs(exact_value - result3)
error_rel_4a = error_abs_4a / abs(exact_value)

print(f"\nError absoluto: {error_abs_4a:.10f}")
print(f"Error relativo: {error_rel_4a:.6e}")
print(f"Error relativo porcentual: {error_rel_4a*100:.4f}%")
```

#### Ejercicio 5: Aproximación de π con serie de Maclaurin

Usar la serie de Maclaurin para arcotangente para aproximar $\pi$.

**Serie:**
$$\arctan(x) \approx x - \frac{x^3}{3} + \frac{x^5}{5}$$

```python
def arctan_approx(x):
    """
    Aproximación de arctan usando los primeros 3 términos de Maclaurin
    """
    return x - (x**3)/3 + (x**5)/5

# Verificar la función
print(f"arctan(0.5) ≈ {arctan_approx(0.5):.10f}")
print(f"arctan real(0.5) = {np.arctan(0.5):.10f}")
```

a) $4\left[\arctan\left(\frac{1}{2}\right) + \arctan\left(\frac{1}{3}\right)\right]$

```python
print("Expresión: 4[arctan(1/2) + arctan(1/3)]\n")

# Calcular cada arctan
arctan_half = arctan_approx(1/2)
arctan_third = arctan_approx(1/3)

print(f"arctan(1/2) ≈ {arctan_half:.10f}")
print(f"arctan(1/3) ≈ {arctan_third:.10f}")

# Aproximación de π
pi_approx_5a = 4 * (arctan_half + arctan_third)
print(f"\nπ ≈ 4 × ({arctan_half:.6f} + {arctan_third:.6f})")
print(f"π ≈ {pi_approx_5a:.10f}")

# Valor real
print(f"π real = {np.pi:.10f}")

# Errores
error_abs_5a = abs(np.pi - pi_approx_5a)
error_rel_5a = error_abs_5a / np.pi

print(f"\nError absoluto: {error_abs_5a:.10f}")
print(f"Error relativo: {error_rel_5a:.6e}")
print(f"Error relativo porcentual: {error_rel_5a*100:.4f}%")
```

```python
print("Expresión: 16*arctan(1/5) - 4*arctan(1/239) [Fórmula de Machin]\n")

# Calcular cada arctan
arctan_fifth = arctan_approx(1/5)
arctan_239 = arctan_approx(1/239)

print(f"arctan(1/5) ≈ {arctan_fifth:.10f}")
print(f"arctan(1/239) ≈ {arctan_239:.10f}")

# Aproximación de π
pi_approx_5b = 16*arctan_fifth - 4*arctan_239
print(f"\nπ ≈ 16 × {arctan_fifth:.6f} - 4 × {arctan_239:.6f}")
print(f"π ≈ {pi_approx_5b:.10f}")

# Valor real
print(f"π real = {np.pi:.10f}")

# Errores
error_abs_5b = abs(np.pi - pi_approx_5b)
error_rel_5b = error_abs_5b / np.pi

print(f"\nError absoluto: {error_abs_5b:.10e}")
print(f"Error relativo: {error_rel_5b:.6e}")
print(f"Error relativo porcentual: {error_rel_5b*100:.6f}%")
```

#### Ejercicio 6: Aproximación de e con series

Aproximar el número de Euler usando la serie:

$$e = \sum_{n=0}^{\infty} \frac{1}{n!}$$

donde $n! = n(n-1) \cdots 2 \cdot 1$ para $n \neq 0$ y $0! = 1$.

a) $\sum_{n=0}^{5} \frac{1}{n!}$

```python
print("Expresión: Σ(1/n!) para n=0 hasta n=5\n")

import math

# Calcular la suma
e_approx_6a = 0
print("Términos de la serie:")
for n in range(6):
    term = 1/math.factorial(n)
    e_approx_6a += term
    print(f"n={n}: 1/{n}! = 1/{math.factorial(n)} = {term:.10f}")

print(f"\nSuma total: e ≈ {e_approx_6a:.10f}")
print(f"e real = {np.e:.10f}")

# Errores
error_abs_6a = abs(np.e - e_approx_6a)
error_rel_6a = error_abs_6a / np.e

print(f"\nError absoluto: {error_abs_6a:.10f}")
print(f"Error relativo: {error_rel_6a:.6e}")
print(f"Error relativo porcentual: {error_rel_6a*100:.4f}%")
```

b) $\sum_{n=0}^{10} \frac{1}{n!}$

```python
print("Expresión: Σ(1/n!) para n=0 hasta n=10\n")

# Calcular la suma
e_approx_6b = 0
print("Términos de la serie:")
for n in range(11):
    term = 1/math.factorial(n)
    e_approx_6b += term
    if n <= 5:
        print(f"n={n}: 1/{n}! = {term:.10f}")
    else:
        print(f"n={n}: 1/{n}! = {term:.15e}")

print(f"\nSuma total: e ≈ {e_approx_6b:.15f}")
print(f"e real = {np.e:.15f}")

# Errores
error_abs_6b = abs(np.e - e_approx_6b)
error_rel_6b = error_abs_6b / np.e

print(f"\nError absoluto: {error_abs_6b:.15e}")
print(f"Error relativo: {error_rel_6b:.15e}")
print(f"Error relativo porcentual: {error_rel_6b*100:.12f}%")
```
