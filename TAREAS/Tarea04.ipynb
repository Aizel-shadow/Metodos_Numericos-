1. Use el método de bisección para encontrar soluciones precisas dentro de $10^{-2}$ para:

a) $x^3 - 7x^2 + 14x - 6 = 0$ en $[0,1]$

b) $x^3 - 7x^2 + 14x - 6 = 0$ en $[1,3.2]$

c) $x^3 - 7x^2 + 14x - 6 = 0$ en $[3.2,4]$

````python
import numpy as np
import matplotlib.pyplot as plt

def biseccion(f, a, b, tol=1e-3, max_iter=100):
    if f(a)*f(b) > 0:
        raise ValueError("No hay cambio de signo en el intervalo.")
    for _ in range(max_iter):
        c = (a + b) / 2
        if abs(f(c)) < tol or abs(b - a) < tol:
            return c
        if f(a)*f(c) < 0:
            b = c
        else:
            a = c
    return c

f = lambda x: x**3 - 7*x**2 + 14*x - 6

intervalos = [(0,1), (1,3.2), (3.2,4)]
for a,b in intervalos:
    r = biseccion(f,a,b)
    print(f"Raíz en [{a},{b}] ≈ {r:.5f}")

x = np.linspace(0,4,400)
plt.axhline(0,color='gray',ls='--')
plt.plot(x,f(x))
plt.title("f(x) = x³ - 7x² + 14x - 6")
plt.xlabel("x"); plt.ylabel("f(x)")
plt.grid(True); plt.show()
````

5.
Sea $f(x) = (x+3)(x+1)^2 x (x-1)^3 (x-3)$

Determinar el cero en cada intervalo:

a) $[-1.5,2.5]$
b) $[-0.5,2.4]$
c) $[-0.5,3]$
d) $[-3,-0.5]$

````python
import numpy as np
import matplotlib.pyplot as plt

def biseccion(f, a, b, tol=1e-6, max_iter=100):
    if f(a) * f(b) > 0:
        raise ValueError(f"No hay cambio de signo en [{a}, {b}] (f(a)={f(a):.4f}, f(b)={f(b):.4f})")
    for _ in range(max_iter):
        c = (a + b) / 2
        if abs(f(c)) < tol or abs(b - a) < tol:
            return c
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return c

# --- EJERCICIO 5 ---
f = lambda x: (x + 3) * (x + 1) * x * (x - 1) * (x - 3)
intervalos = [(-1.5, 2.5), (-0.5, 2.4), (-0.5, 3), (-3, -0.5)]

for a, b in intervalos:
    try:
        r = biseccion(f, a, b)
        print(f"Intervalo [{a}, {b}] → Raíz ≈ {r:.4f}")
    except Exception as e:
        print(f"Intervalo [{a}, {b}] → {e}")

# Graficar función
x = np.linspace(-3.5, 3.5, 400)
plt.plot(x, f(x))
plt.axhline(0, color='gray', ls='--')
plt.title("f(x) = (x+3)(x+1)x(x−1)(x−3)")
plt.grid(True)
plt.show()

````

Un abrevadero de longitud $L$ tiene sección transversal semicircular de radio $r$.
El volumen de agua es:
$$
V = L \Big( 0.5 \pi r^2 - r^2 \arcsin\Big(\frac{h}{r}\Big) - h \sqrt{r^2 - h^2} \Big)
$$

Datos:
$$
L = 10\ \text{cm}, \quad r = 1\ \text{cm}, \quad V = 12.4\ \text{cm}^3
$$

Encontrar:
$$
h \text{ con precisión } 0.01\ \text{cm}

2. Un objeto que cae verticalmente a través del aire está sujeto a una resistencia viscosa, así como a la fuerza
de gravedad. Suponga que un objeto con masa 𝑚 cae desde una altura $𝑠^2$ y que la altura del objeto después
de 𝑡 segundos es 

$$
s(t) = S_0 - \frac{mg}{k} + \frac{m^2 g}{k^2} \Big( 1 - e^{-\frac{k t}{m}} \Big)
$$

Datos:
$$
S_0 = 300\ \text{m}, \quad m = 0.25\ \text{kg}, \quad k = 0.1\ \text{N}\cdot\text{s/m}, \quad g = 9.81\ \text{m/s}^2
$$

Encontrar:
$$
t \text{ cuando } s(t) = 0 \text{ con precisión } 0.01\ \text{s}
$$


````python
import math
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
S0, m, k, g = 300.0, 0.25, 0.1, 9.81

# Función correcta (con m^2)
def s_correct(t):
    return S0 - (m*g/k)*t + (m*m*g/k**2)*(1 - math.exp(-k*t/m))

# Función incorrecta (sin el segundo m)
def s_incorrect(t):
    return S0 - (m*g/k)*t + (m*g/k**2)*(1 - math.exp(-k*t/m))

# Bisección simple
def bisection(f, a, b, tol=1e-6, max_iter=1000):
    fa, fb = f(a), f(b)
    if fa*fb > 0:
        raise ValueError("No hay cambio de signo en el intervalo")
    for _ in range(max_iter):
        c = (a+b)/2
        fc = f(c)
        if abs(fc) < tol or (b-a)/2 < tol:
            return c
        if fa*fc < 0:
            b, fb = c, fc
        else:
            a, fa = c, fc
    return c

# Encontrar raíces
root_correct = bisection(s_correct, 0, 30, 1e-6)
root_incorrect = bisection(s_incorrect, 0, 100, 1e-6)

print(f"Raíz (correcta, m^2): {root_correct:.6f} s")    # ≈ 14.7255
print(f"Raíz (incorrecta, m): {root_incorrect:.6f} s") # ≈ 22.2310
````

````python
# Graficar para comparar
t = np.linspace(0, 30, 400)
plt.plot(t, [s_correct(tt) for tt in t], label='s_correct (m^2)')
plt.plot(t, [s_incorrect(tt) for tt in t], label='s_incorrect (m)')
plt.axhline(0, color='k', ls='--')
plt.legend(); plt.xlabel('t [s]'); plt.ylabel('s(t) [m]'); plt.grid(True)
plt.ylim(-50, 350)
plt.show()
````
