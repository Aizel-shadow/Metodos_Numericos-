1. Utilizar aritmética de corte (truncamiento) de 3 dígitos para calcular sumas. Comparar el orden de suma: ascendente vs descendente.

a) $\sum_{i=1}^{10} \frac{1}{i^2}$

Calcular primero: $\frac{1}{1} + \frac{1}{4} + \cdots + \frac{1}{100}$ (ascendente)

Luego: $\frac{1}{100} + \frac{1}{81} + \cdots + \frac{1}{1}$ (descendente)
```

````python
import numpy as np

def truncar(x, cifras=3):
    """
    Función para truncar (cortar) a n cifras significativas
    NO redondea, solo corta los dígitos
    """
    if x == 0:
        return 0
    from math import log10, floor
    decimales = -int(floor(log10(abs(x)))) + (cifras - 1)
    factor = 10 ** decimales
    return floor(x * factor) / factor

# MÉTODO ASCENDENTE: 1/1 + 1/4 + ... + 1/100
print("\n MÉTODO ASCENDENTE ")
suma_asc = 0
print(f"{'i':<5} {'1/i²':<15} {'fl(1/i²)':<15} {'Suma parcial':<15}")
print("-"*60)

for i in range(1, 11):
    termino_exacto = 1 / (i**2)
    termino_truncado = truncar(termino_exacto, 3)
    suma_asc = truncar(suma_asc + termino_truncado, 3)
    print(f"{i:<5} {termino_exacto:<15.10f} {termino_truncado:<15} {suma_asc:<15}")

print(f"\nResultado ascendente: {suma_asc}")
````

````python
# MÉTODO DESCENDENTE: 1/100 + 1/81 + ... + 1/1
print("\n MÉTODO DESCENDENTE ")
suma_desc = 0
print(f"{'i':<5} {'1/i²':<15} {'fl(1/i²)':<15} {'Suma parcial':<15}")
print("-"*60)

for i in range(10, 0, -1):
    termino_exacto = 1 / (i**2)
    termino_truncado = truncar(termino_exacto, 3)
    suma_desc = truncar(suma_desc + termino_truncado, 3)
    print(f"{i:<5} {termino_exacto:<15.10f} {termino_truncado:<15} {suma_desc:<15}")

print(f"\nResultado descendente: {suma_desc}")
````

````python
# VALOR EXACTO
valor_exacto_1a = sum([1/(i**2) for i in range(1, 11)])
print(f"\n{'-'*60}")
print(f"Valor exacto: {valor_exacto_1a:.10f}")
print(f"Error ascendente:  {abs(valor_exacto_1a - suma_asc):.10f}")
print(f"Error descendente: {abs(valor_exacto_1a - suma_desc):.10f}")

if abs(valor_exacto_1a - suma_desc) < abs(valor_exacto_1a - suma_asc):
    print("\n✓ MÉTODO DESCENDENTE ES MÁS PRECISO")
else:
    print("\n✓ MÉTODO ASCENDENTE ES MÁS PRECISO")
````
