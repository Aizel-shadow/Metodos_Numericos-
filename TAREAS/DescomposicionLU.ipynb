1. Multiplicación Matriz-Matriz

```python
import numpy as np

# 1a
A1a = np.array([[2, -3], [3, -1]])
B1a = np.array([[1, 5], [2, 0]])
print("1a:\n", A1a @ B1a)

# 1b
A1b = np.array([[2, -3], [3, -1]])
B1b = np.array([[1, 5, -4], [-3, 2, 0]])
print("\n1b:\n", A1b @ B1b)

# 1c
A1c = np.array([[2, -3, 1], [4, 3, 0], [5, 2, -4]])
B1c = np.array([[0, 1, -2], [1, 0, -1], [2, 3, -2]])
print("\n1c:\n", A1c @ B1c)

# 1d
A1d = np.array([[2, 1, 2], [-2, 3, 0], [2, -1, 3]])
B1d = np.array([[1, -2], [-4, 1], [0, 2]])
print("\n1d:\n", A1d @ B1d)
```

2. Matrices No Singulares e Inversas

```python
def procesar_2(M, nombre):
    det = np.linalg.det(M)
    print(f"\n--- Matriz {nombre} ---")
    if not np.isclose(det, 0):
        print("Es NO singular. Inversa:\n", np.linalg.inv(M))
    else:
        print("Es SINGULAR (No tiene inversa).")

procesar_2(np.array([[4, 2, 6], [3, 0, 7], [-2, -1, -3]]), "2a")
procesar_2(np.array([[1, 2, 0], [2, 1, -1], [3, 1, 1]]), "2b")
procesar_2(np.array([[1, 1, -1, 1], [1, 2, -4, -2], [2, 1, 1, 5], [-1, 0, -2, -4]]), "2c")
procesar_2(np.array([[4, 0, 0, 0], [6, 7, 0, 0], [9, 11, 1, 0], [5, 4, 1, 1]]), "2d")
```

4. Valor de $\alpha$ para Matriz Singular

```python
import sympy as sp
a_sym = sp.symbols('alpha')
A4 = sp.Matrix([[1, -1, a_sym], [2, 2, 1], [0, a_sym, -1.5]])
sol_alpha = sp.solve(A4.det(), a_sym)
print("Valores de alpha para que A sea singular:", sol_alpha)
```

5. Resolución de Sistemas $LUx = b$

```python
def resolver_lu_manual(L, U, b):
    y = np.linalg.solve(L, b)
    x = np.linalg.solve(U, y)
    return x

# 5a
L5a, U5a = np.array([[1,0,0],[2,1,0],[-1,0,1]]), np.array([[2,3,-1],[0,-2,1],[0,0,3]])
b5a = np.array([2, -1, 1])
print("Solución 5a:", resolver_lu_manual(L5a, U5a, b5a))

# 5b
L5b, U5b = np.array([[2,0,0],[-1,1,0],[3,2,-1]]), np.array([[1,1,1],[0,1,2],[0,0,1]])
b5b = np.array([-1, 3, 0])
print("Solución 5b:", resolver_lu_manual(L5b, U5b, b5b))
```

6. Factorización LU

```python
from scipy.linalg import lu

def mostrar_lu(A, nombre):
    P, L, U = lu(A)
    print(f"\n--- Factorización LU {nombre} ---")
    print("L:\n", L, "\nU:\n", U)

# Matrices de los ejercicios
mostrar_lu(np.array([[2, -1, 1], [3, 3, 9], [3, 3, 5]]), "6a")
mostrar_lu(np.array([[1.012, -2.132, 3.104], [-2.132, 4.096, -7.013], [3.104, -7.013, 0.014]]), "6b")
mostrar_lu(np.array([[2,0,0,0],[1,1.5,0,0],[0,-3,0.5,0],[2,-2,1,1]]), "6c")
mostrar_lu(np.array([[2.1756, 4.0231, -2.1732, 5.1967], [-4.0231, 6.0000, 0, 1.1973], 
                     [-1.0000, -5.2107, 1.1111, 0], [6.0235, 7.0000, 0, -4.1561]]), "6d")
```

7. Resolución de Sistemas usando LU

```python
def resolver_punto_7(A, b, nombre):
    P, L, U = lu(A)
    y = np.linalg.solve(L, P.T @ b)
    x = np.linalg.solve(U, y)
    print(f"Resultado {nombre}:", x)

# 7a
resolver_punto_7(np.array([[2,-1,1],[3,3,9],[3,3,5]]), np.array([-1,0,4]), "7a")
# 7b
resolver_punto_7(np.array([[1.012, -2.132, 3.104], [-2.132, 4.096, -7.013], [3.104, -7.013, 0.014]]), 
                 np.array([1.984, -5.049, -3.895]), "7b")
# 7c
resolver_punto_7(np.array([[2,0,0,0],[1,1.5,0,0],[0,-3,0.5,0],[2,-2,1,1]]), 
                 np.array([3, 4.5, -6.6, 0.8]), "7c")
# 7d
resolver_punto_7(np.array([[2.1756, 4.0231, -2.1732, 5.1967], [-4.0231, 6.0000, 0, 1.1973], 
                           [-1.0000, -5.2107, 1.1111, 0], [6.0235, 7.0000, 0, -4.1561]]), 
                 np.array([17.102, -6.1593, 3.0004, 0.0000]), "7d")
```
