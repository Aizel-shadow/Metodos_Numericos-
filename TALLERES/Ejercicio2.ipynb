```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton

# Definir la función y su derivada
def f(x):
    return x**3 - 3 * x**2 + x - 1

def fprime(x):
    return 3 * x**2 - 6 * x + 1

# Aplicar el método de Newton-Raphson
x0 = 10
raiz = newton(f, x0=x0, fprime=fprime)

# Mostrar resultado
print("=" * 50)
print("MÉTODO DE NEWTON-RAPHSON")
print("=" * 50)
print(f"Punto inicial: x₀ = {x0}")
print(f"Raíz aproximada: x = {raiz:.10f}")
print(f"Verificación: f({raiz:.6f}) = {f(raiz):.2e}")
print("=" * 50)

# Crear la gráfica
x_vals = np.linspace(-1, 4, 400)
y_vals = f(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, 'b-', linewidth=2, label='f(x) = x³ - 3x² + x - 1')
plt.axhline(y=0, color='k', linestyle='--', linewidth=0.8, alpha=0.5)
plt.axvline(x=0, color='k', linestyle='--', linewidth=0.8, alpha=0.5)
plt.plot(raiz, f(raiz), 'ro', markersize=12, label=f'Raíz ≈ {raiz:.6f}')
plt.plot(x0, f(x0), 'go', markersize=10, label=f'Punto inicial x₀ = {x0}')
plt.grid(True, alpha=0.3)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title('Método de Newton-Raphson', fontsize=14, fontweight='bold')
plt.legend(fontsize=11)
plt.ylim([-10, 20])
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Definir la función y su derivada
def f(x):
    return x**3 - 3 * x**2 + x - 1

def fprime(x):
    return 3 * x**2 - 6 * x + 1

# Método de Newton-Raphson con registro de iteraciones
def newton_tabla(f, fprime, x0, tol=1e-8, max_iter=50):
    datos = []
    x = x0
    
    for i in range(max_iter):
        fx = f(x)
        fpx = fprime(x)
        error = abs(fx)
        
        datos.append({
            'n': i,
            'x_n': x,
            'f(x_n)': fx,
            'f_prima(x_n)': fpx,
            'Error': error
        })
        
        if error < tol:
            break
        
        x = x - fx / fpx
    
    return x, pd.DataFrame(datos)

# Ejecutar el método
x0 = 10
raiz, tabla = newton_tabla(f, fprime, x0)

# Mostrar tabla
print("\n" + "="*90)
print("TABLA DE ITERACIONES - MÉTODO DE NEWTON-RAPHSON")
print("="*90)
print(f"Función: f(x) = x³ - 3x² + x - 1")
print(f"Punto inicial: x₀ = {x0}")
print("="*90 + "\n")

# Configurar pandas para mejor visualización
pd.set_option('display.float_format', lambda x: f'{x:.10f}' if abs(x) > 1e-3 else f'{x:.10e}')
print(tabla.to_string(index=False))

print("\n" + "="*90)
print(f"Raíz aproximada: x = {raiz:.10f}")
print(f"Verificación: f({raiz:.10f}) = {f(raiz):.2e}")
print(f"Iteraciones totales: {len(tabla)}")
print("="*90)
```
