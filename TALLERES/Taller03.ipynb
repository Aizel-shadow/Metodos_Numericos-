3. Gráficar $\frac{1}{1-x}$

3.1. Gráficar Primer Orden

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-0.5, 1.4, 500)
x0 = 0.5  # Punto de expansión

P1 = 2 + 4*(x - x0)

plt.figure()
plt.plot(x, 1/(1-x), 'k-', label='1/(1-x)')
plt.plot(x, P1, 'b--', label='Orden 1')
plt.plot(x0, 2, 'ro', markersize=8, label='x₀=0.5')
plt.grid(True)
plt.legend()
plt.title('Orden 1: P₁(x) = 2 + 4(x-0.5)')
plt.ylim(0, 6)
plt.show()
```

---

3.2. Gráficar Segundo Orden

```python
P2 = 2 + 4*(x - x0) + 8*(x - x0)**2

plt.figure()
plt.plot(x, 1/(1-x), 'k-', label='1/(1-x)')
plt.plot(x, P2, 'g--', label='Orden 2')
plt.plot(x0, 2, 'ro', markersize=8, label='x₀=0.5')
plt.grid(True)
plt.legend()
plt.title('Orden 2: P₂(x) = 2 + 4(x-0.5) + 8(x-0.5)²')
plt.ylim(0, 6)
plt.show()
```

---

3.3. Gráficar Tercer Orden

```python
P3 = 2 + 4*(x - x0) + 8*(x - x0)**2 + 16*(x - x0)**3

plt.figure()
plt.plot(x, 1/(1-x), 'k-', label='1/(1-x)')
plt.plot(x, P3, 'r--', label='Orden 3')
plt.plot(x0, 2, 'ro', markersize=8, label='x₀=0.5')
plt.grid(True)
plt.legend()
plt.title('Orden 3: P₃(x) = 2 + 4(x-0.5) + 8(x-0.5)² + 16(x-0.5)³')
plt.ylim(0, 6)
plt.show()
```

### Polinomio de Lagrange 


---

1. $$ (0,0),(30,0.5),(60,\frac{√3}{2}),(90,1)$$


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = np.array([0, 30, 60, 90])
y_data = np.array([0, 0.5, np.sqrt(3)/2, 1])

# Función para calcular L_k(x)
def L_k(x, x_data, k):
    n = len(x_data)
    L = 1
    for i in range(n):
        if i != k:
            L *= (x - x_data[i]) / (x_data[k] - x_data[i])
    return L

# Polinomio de Lagrange
def P(x, x_data, y_data):
    n = len(x_data)
    resultado = 0
    for k in range(n):
        resultado += y_data[k] * L_k(x, x_data, k)
    return resultado
```

```python
x = np.linspace(-10, 100, 500)
y = [P(xi, x_data, y_data) for xi in x]

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2, label='P(x) Lagrange')
plt.plot(x_data, y_data, 'ro', markersize=10, label='Puntos dados')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ejercicio 1: Interpolación de Lagrange')
plt.legend()
plt.show()
```

2. $$(1,1),(2,2),(3,2)$$

```python
x_data = np.array([1, 2, 3])
y_data = np.array([1, 2, 2])

x = np.linspace(0, 4, 500)
y = [P(xi, x_data, y_data) for xi in x]

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'g-', linewidth=2, label='P(x) Lagrange')
plt.plot(x_data, y_data, 'ro', markersize=10, label='Puntos dados')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ejercicio 2: Interpolación de Lagrange')
plt.legend()
plt.show()
```
