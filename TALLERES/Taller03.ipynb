1. Gráficar In(x)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.1, 5, 500)

plt.plot(x, np.log(x), 'b-', linewidth=2)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('ln(x)')
plt.show()
```

---


2. Gráficar Primer Orden 

```python
x = np.linspace(0.1, 3, 500)

plt.figure()
plt.plot(x, np.log(x), 'k-', label='ln(x)')
plt.plot(x, x-1, 'b--', label='Orden 1')
plt.grid(True)
plt.legend()
plt.title('Orden 1')
plt.show()
```

---

3. Gráficar Segundo Orden

```python
plt.figure()
plt.plot(x, np.log(x), 'k-', label='ln(x)')
plt.plot(x, (x-1) - 0.5*(x-1)**2, 'g--', label='Orden 2')
plt.grid(True)
plt.legend()
plt.title('Orden 2')
plt.show()
```

---

4. Gráficar Tercer Orden 

```python
plt.figure()
plt.plot(x, np.log(x), 'k-', label='ln(x)')
plt.plot(x, (x-1) - 0.5*(x-1)**2 + (1/3)*(x-1)**3, 'r--', label='Orden 3')
plt.grid(True)
plt.legend()
plt.title('Orden 3')
plt.show()
```
