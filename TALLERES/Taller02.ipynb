**1.** Encuentre todas las raíces del polinomio:

$$
x^{4} + 540x^{3} + 109124x^{2} + 9781632x + 328188672 = 0
$$




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton, bisect


def f1(x):
    return x**4 + 540*x**3 + 109124*x**2 + 9781632*x + 328188672

def f1_prime(x):
    return 4*x**3 + 1620*x**2 + 218248*x + 9781632


# Graficar para visualizar raíces
x_vals = np.linspace(-400, 50, 1000)
y_vals = f1(x_vals)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, 'b-', linewidth=2)
plt.axhline(y=0, color='r', linestyle='--', alpha=0.5)
plt.grid(True, alpha=0.3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Ejercicio 1: Polinomio grado 4')

# Buscar cambios de signo para bisección
raices_1 = []
intervalos = [(-400, -200), (-200, -100), (-100, -50), (-50, 0)]

for a, b in intervalos:
    if f1(a) * f1(b) < 0:
        raiz = bisect(f1, a, b)
        raices_1.append(raiz)
        plt.plot(raiz, 0, 'ro', markersize=8)
        print(f"Raíz encontrada (bisección): x = {raiz:.6f}")

plt.legend(['f(x)', 'y=0', 'Raíces'])
plt.show()

print(f"\nTotal de raíces encontradas: {len(raices_1)}\n")
```

**2.** Encuentre todos los puntos en los que la curva

$$
\left(\frac{y}{2}\right)^{2} = \left(\frac{x + 3}{2}\right)^{2} - 1
$$

interseca el eje \( y = -2 \)


```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton, bisect

# Sustituir y = -2 en la ecuación
# (-2/2)^2 = ((x+3)/2)^2 - 1
# 1 = ((x+3)/2)^2 - 1
# ((x+3)/2)^2 = 2

def f2(x):
    return ((x + 3) / 2)**2 - 2 - 1

def f2_prime(x):
    return (x + 3) / 2

# Graficar
x_vals2 = np.linspace(-10, 5, 1000)
y_curve = np.sqrt(2 * ((x_vals2 + 3) / 2)**2 - 2) * 2  # Despejando y
y_line = -2 * np.ones_like(x_vals2)

plt.figure(figsize=(10, 5))
plt.plot(x_vals2, y_curve, 'b-', label='Curva', linewidth=2)
plt.plot(x_vals2, -y_curve, 'b-', linewidth=2)
plt.plot(x_vals2, y_line, 'r-', label='y = -2', linewidth=2)
plt.grid(True, alpha=0.3)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ejercicio 2: Intersección de curvas')

# Encontrar puntos de intersección
puntos_interseccion = []
x_inicial = [-5, 0]

for x0 in x_inicial:
    try:
        raiz = newton(f2, x0)
        puntos_interseccion.append(raiz)
        plt.plot(raiz, -2, 'go', markersize=10)
        print(f"Punto de intersección: x = {raiz:.6f}, y = -2")
    except:
        pass

plt.legend()
plt.show()

print(f"\nTotal de puntos: {len(puntos_interseccion)}\n")
```

**3.** Dada la función 

$$
f(x) = \frac{\sin(x)}{x}
$$

¿A partir de qué valor \( x_T \) se cumple que 

$$
f(x) < 0.015, \quad \forall x \ge x_T ?
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton, bisect

def f3(x):
    if x == 0:
        return 1
    return np.sin(x) / x - 0.015

def f3_prime(x):
    if abs(x) < 1e-10:
        return 0
    return (np.cos(x) * x - np.sin(x)) / x**2

# Graficar
x_vals3 = np.linspace(0.1, 50, 1000)
y_vals3 = np.sin(x_vals3) / x_vals3

plt.figure(figsize=(10, 5))
plt.plot(x_vals3, y_vals3, 'b-', linewidth=2, label='sin(x)/x')
plt.axhline(y=0.015, color='r', linestyle='--', alpha=0.7, label='y = 0.015')
plt.grid(True, alpha=0.3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Ejercicio 3: sin(x)/x')

# Encontrar x_T donde f(x) = 0.015
try:
    x_T = newton(f3, 10)
    plt.plot(x_T, 0.015, 'go', markersize=10)
    print(f"x_T = {x_T:.6f}")
    print(f"Para x ≥ {x_T:.6f}, se cumple que f(x) < 0.015")
    plt.axvline(x=x_T, color='g', linestyle=':', alpha=0.5, label=f'x_T = {x_T:.2f}')
except Exception as e:
    print(f"Error: {e}")

plt.legend()
plt.show()

```
