¿Qué representaciones de datos soportan estas GPUs (FP64, FP32, INT8)?

Ambas arquitecturas son versátiles, pero Blackwell introduce niveles de eficiencia que Hopper no tenía:

* FP64 (Precisión doble): Soportado por ambas, es vital para simulaciones científicas (clima, física cuántica) donde el error decimal no puede existir.

* FP32: Soportado para tareas generales y entrenamiento estándar.

* INT8: Soportado ampliamente para inferencia (ejecutar modelos ya entrenados), reduciendo el consumo de memoria.

* FP8 / FP4: Aquí está el salto. Hopper introdujo FP8. Blackwell introduce soporte nativo para FP4, permitiendo que modelos el doble de grandes quepan en la misma memoria.


¿Por qué la nueva arquitectura prefiere representaciones numéricas con menor precisión?

1. Ancho de banda y Memoria: Un número en FP32 ocupa 4 bytes, un número en FP4 ocupa solo 0.5 bytes, esto significa que puedes mover 8 veces más datos a través de la memoria en el mismo tiempo.

2. Eficiencia Energética: Multiplicar números pequeños (de pocos bits) requiere muchísimos menos transistores y, por ende, consume mucha menos electricidad y genera menos calor.

3. Resiliencia de las Redes Neuronales: Se ha descubierto que las redes neuronales son "ruidosas" por naturaleza. No necesitan 7 decimales de precisión para saber si una imagen es un gato o un perro; les basta con la dirección general del valor. La cuantización (bajar la precisión) permite modelos más rápidos con una pérdida de exactitud casi imperceptible.
