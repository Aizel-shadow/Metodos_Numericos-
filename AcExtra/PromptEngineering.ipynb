### 1. Usar Delimitadores Claros

La IA puede confundirse cuando se mezcla instrucciones con el texto que quieres que analice. El uso de delimitadores como `"""`, `---`, `###` o etiquetas tipo XML (`<texto> ... </texto>`) para separar secciones.

* **Ejemplo:** *"Resume el siguiente texto delimitado por triples comillas: """[Tu texto aquí]""""*

### 2. Especificar el Formato de Salida (Output)

No dejar que la IA decida cómo entregar la información. Ser extremadamente específico sobre la estructura.

* **Pide formatos específicos:** *"Entrégame la respuesta en una tabla de 3 columnas"*, *"Genera un archivo JSON"*, o *"Usa viñetas y no escribas más de 50 palabras"*.

### 3. Implementar el "Prompt Negativo"

A veces es más importante decir qué **no** quieres que decir qué quieres. Esto evita muletillas comunes de la IA o formatos indeseados.

* **Ejemplo:** *"Explica cómo funciona un motor de combustión, pero **no** menciones a Tesla, **no** uses lenguaje técnico complejo y **no** incluyas una introducción larga; ve directo al grano"*.

### 4. La Técnica del "Prompt de Autocrítica"

Antes de dar por buena una respuesta, pide a la IA que revise su propio trabajo. Esto eleva la calidad exponencialmente.

* **Instrucción:** *"Después de escribir el artículo, actúa como un editor crítico, encuentra 3 puntos débiles en tu argumentación y reescríbelo para corregirlos"*.

### 5. Tiempo para "Pensar"

Si se hace una pregunta compleja y se pide una respuesta inmediata, la IA puede cometer errores de cálculo o lógica.

* **Truco:** Pedirle que primero enumere las variables involucradas o que haga un esquema de la solución antes de escribir la respuesta definitiva. Esto se conoce como *pre-computing* dentro del prompt.
