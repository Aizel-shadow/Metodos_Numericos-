#### ***SESGOS COGNITIVOS***

*1. Sesgo de Confirmación*

Es cuando buscamos o interpretamos información que confirma lo que ya creemos.

*Ejemplo*

Si se cree que los gatos son mejores que los perros, solo se recordara las veces que los perros ladraron molestamente, pero se ignorara cuando un gato arañó a alguien. 

*2. Efecto Halo*

Es cuando una característica positiva de alguien influye en cómo vemos sus otras cualidades.

*Ejemplo*

Supongamos que tu profesor favorito hace un chiste malo, pero te parece gracioso solo porque te cae bien.

*3. Sesgo de Anclaje*

Es cuando nos basamos demasiado en la primera información que recibimos. 

*Ejemplo*

Una camisa tiene un precio "original" de $100 tachado y ahora cuesta $50. Piensas que es una gran oferta, aunque quizá siempre valió $50, pero el primer número "ancló" tu percepción.

*4. Sesgo de Disponibilidad*

Sobrestimamos la probabilidad de eventos que recordamos fácilmente. 

*Ejemplo*

Después de ver noticias sobre accidentes de avión, piensas que volar es muy peligroso, aunque estadísticamente es más seguro que conducir un auto.

*5. Efecto Dunning-Kruger*

Personas con poco conocimiento sobreestiman su competencia. 

*Ejemplo*

Alguien ve un video de 10 minutos sobre economía y cree que puede dar consejos mejor que un economista con años de estudio.

#### ***FALACIAS LÓGICAS***

*1. Ad Hominem*

Atacar a la persona en lugar de su argumento. 

*Ejemplo*

"No puedes hablar sobre un circuito electrico porque reprobaste electronica", en lugar de refutar el argumento, atacas a la persona.

*2. Falacia del Hombre de Paja*

Distorsionar el argumento del oponente para que sea más fácil atacarlo.

*Ejemplo*

Persona A: "Deberían arreglar los baños de una vez"

Persona B: "¿Entonces prefieres que cierren la facultad todo el semestre 
para remodelar y tengamos que estar virtuales?

*3. Falso Dilema*

Presentar solo dos opciones cuando hay más.

*Ejemplo*

"O estás conmigo o estás contra mí", ignorando que alguien puede ser neutral o tener una posición intermedia.

*4. Pendiente Resbaladiza*

Asumir que una acción llevará inevitablemente a consecuencias extremas.

*Ejemplo*

"Si permitimos que los estudiantes usen celulares en clase, pronto nadie prestará atención, todos reprobarán y la educación colapsará."

*5. Apelación a la Tradición*

Argumentar que algo es correcto solo porque "siempre se ha hehco así".

*Ejemplo*

"Siempre hemos comido a las 12, ¿por qué cambiar ahora?", sin considerar si hay mejores razones para hacerlo diferente.
