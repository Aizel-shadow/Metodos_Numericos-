1. Investigar sobre los atajos de teclado para ingresar ecuaciones matemáticas en Word y Latex. 

*1.1. WORD*

* Al presionar ```Alt + =``` se activara el editor de ecuaciones.
* Se escribe el comando (como ```\alpha```) y luego se presiona ```Espacio``` para convertirlo al símbolo.
* Los comandos son similares a LaTeX.

*1.2. LaTeX*

* Usa delimitadores matemáticos ```$...$``` para *inline* o ```$$...$$``` para ecuaciones centradas. 
* Los comandos llevan la barra invertidad (```\```) seguida del nombre del símbolo.
* Es muy utilizado en documentos científicos y académicos.

*1.3. Jupyter Notebook*

* Cambia la celda a tipo "Markdown".
* Se usa la sintaxis de LaTeX dentro de delimitadores ```$...$```.
* Es ideal para documentar código con ecuaciones matemáticas.
